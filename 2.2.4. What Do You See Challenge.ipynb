# Global Rates of Adult Obesity 

*Data source: http://data.un.org/Data.aspx?d=GenderStat&f=inID%3a43*

This notebook explores a United Nations data set that looks at the global rate of obesity in adult populations in various countries.  For a given country, the percentage of the population classified as obese is reported and differentiated by gender.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#Import data and create data frame.
obesity = pd.read_csv('C:/Users/Doug/OneDrive/Python/data_files/UNdata_obesity.csv')
obesity.columns = ['country', 'gender', 'year', 'source', 'unit', 'value']
```

```python
#Create data for gender counts.
gender_count = obesity.gender.value_counts().tolist()

plt.figure(figsize=(10,5))
plt.pie(gender_count, startangle=90, colors=('lightcoral', 'lightskyblue'), labels=('Female', 'Male'), autopct='%1.0f%%')
plt.axis('equal')
plt.show()
```

Using a pie chart to look at the preliminary distribution of data between male and female subgroups, we see that 40% of the  data represents males and 60% of the data represents females.

```python
#Create data frame for females. 
females=obesity[obesity['gender'] == 'Female 15+ yr']

#Create data frame for males. 
males=obesity[obesity['gender'] == 'Male 15+ yr']
```

```python
#Look at distribution of data based on total population

plt.hist(obesity['value'], bins=20, color='grey')
plt.title('Obesity Rates')
plt.xlabel('Percent Obesity')
plt.ylabel('count')
plt.show()
```

Now let's generate a histogram on the entire data set to take a closer look at the distribution of data.  The most frequently observed obesity rates range from 0 to ~25%.  The data set has several outliers representing high obesity rates.

```python
#Compare total distribution of male obesity and female obesity.
plt.hist(males['value'], normed=True, bins=20, color='lightskyblue', alpha=0.4)
plt.hist(females['value'], normed=True, bins=20, color='lightcoral', alpha=0.4)
plt.title('Normalized Obesity Rates By Gender')
plt.xlabel('Percent Obesity')
plt.ylabel('Normalized Count')
plt.show()
```

Taking it a step further, let's look at how the data varies based on gender.  In the above histogram, males are represented by blue and females are represented by pink.  Due to the unequal samples size among male and female data sets, results have been normalized.  Females show a wider distribution in obesity rates relative to males.  Both male and female groups contain several outliers with high obesity rates.  Both groups are also skewed to the left, which makes sense considering this data represents a sub-population of all adults.

```python
#Generate box and whiskers plot for total population, male, and female data sets.
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.boxplot(males['value'])
plt.title('Males')
plt.ylabel('Percent Obesity')

plt.subplot(1,2,2)
plt.boxplot(females['value'])
plt.title('Females')

plt.tight_layout()
plt.show()
```

Let's take a deeper dive into the data.  The box graphs above illustrate that the median rate of obesity is similar for both male and females, however, females show a greater variety in obesity rates.  Additionally, the female group has more and larger outliers.  

```python
plt.scatter(
    x=obesity['year'],
    y=obesity['value'],
    marker='o',
    color='g'
)
plt.title('Variation in Obesity Rates')
plt.ylabel('Percent Obesity')

plt.show()
```

The scatter plot above shows a distribution of all the data based on year. Multiple data points exist for each year because obesity rates are reported as a discrete interval (grouped by year).  From this plot, we can see a wide variation in obesity rates were observed each year.
